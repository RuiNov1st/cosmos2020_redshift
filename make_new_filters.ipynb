```python
import eazy
import os
import numpy as np
import pandas as pd
```

### CFHT/MegaPrime u*

```python
res_file = '/data/home/wsr/WorkSpace/EAZY/cosmos2020-readcat/COSMOS2020_Farmer_EAZY/Filters/sources/u_megaprime_sagem.res'
with open(res_file,'r') as f1:
    lines = f1.readlines()
wave = []
trans = []
for line in lines:
    lsp = line.strip().split()
    wave.append(float(lsp[0]))
    trans.append(float(lsp[1]))
wave = np.array(wave)
trans = np.array(trans)
u_star_fil = eazy.filters.FilterDefinition(wave=wave,throughput=trans,name='CFHT/MegaPrime u*')
```

```python
u_star_fil.for_filter_file()
```

```python
# write
u_star_fil.write_filter_file(file='/data/home/wsr/WorkSpace/EAZY/cosmos2020-readcat/COSMOS2020_Farmer_EAZY/Filters/u_megaprime_sagem.res')
```
